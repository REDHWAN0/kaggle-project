```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/kaggle/input/taylor-swift-spotify-dataset/taylor_swift_spotify.csv")
print("df shap:", df.shape)
print(df.info())
df.head()
```

```python
df.sort_values("popularity", ascending=True).head(10)
```

```python
mask = df["popularity"] > 0
df[mask].sort_values("popularity", ascending=True).head(10)
```

```python
plt.scatter(df[mask]["danceability"], df[mask]["popularity"])
plt.xlabel("Danceability")
plt.ylabel("Popularity")
plt.title("Tylor Swift: Danceability vs Popularity")
```
